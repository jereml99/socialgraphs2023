```python
import requests
import os
import pandas as pd

df_rappers = pd.read_csv("../../files/Rappers.csv")
baseurl = "http://en.wikipedia.org/w/api.php?"

for rapper in df_rappers["WikipediaPageName"]:
    file_path = f"../data/clean/{rapper}.txt"
    
    # Check if the file already exists
    if os.path.exists(file_path):
        #print(f"File {rapper}.txt already exists. Skipping.")
        continue

    params = {
        "action": "query",
        "prop": "extracts",
        "exlimit": "1",
        "explaintext": "1",
        "format": "json",
        "titles": rapper
     }
    wikitext = requests.get(baseurl, params = params)
    wikijson = wikitext.json()
    num = list(wikijson['query']['pages'].keys())[0]
    text = wikijson['query']['pages'][num]['extract']
    try:
        with open(f"../data/clean/{rapper}.txt", "w", encoding="utf-8") as f:
            f.write(text)
    except Exception as e:
        print(f"Failed to save the page: {rapper}.")
```

> * Use `.index()` to find the index of the word *Snoop*. You'll need to insert this word as an argument between the parentheses. By a process of trial and error, find the slice for the complete sentence that contains this word.


```python
# Find the index of the word 'Snoop'
index = words.index('Snoop')

# Find the start and end indices of the sentence containing 'Snoop'
start_index = index
while start_index > 0 and words[start_index - 1] != '.':
    start_index -= 1
end_index = index
while end_index < len(words) - 1 and words[end_index + 1] != '.':
    end_index += 1

# Convert the list of words in the sentence back to a string
sentence = ' '.join(words[start_index:end_index + 1])
print(sentence + ".")
```

```python
from nltk.corpus.reader import PlaintextCorpusReader

corpus_root = '../data/clean/'

# Create separate lists for the fileids of each coast
fileids_west = []
fileids_east = []

for rapper in df_rappers.to_dict('records'):
    rappertxt = rapper["WikipediaPageName"] + ".txt"

    if(rappertxt == "K-the-I???.txt"):
        rappertxt = "K-the-I.txt"

    if rapper["Coast"] == "West":
        fileids_west.append(rappertxt)
    elif rapper["Coast"] == "East":
        fileids_east.append(rappertxt)

# Initialize PlaintextCorpusReader with the lists of fileids
corpusw = PlaintextCorpusReader(corpus_root, fileids_west)
corpuse = PlaintextCorpusReader(corpus_root, fileids_east)

# Extract the words, sentences, and paragraphs from each corpus
wordsw = corpusw.words()
sentencesw = corpusw.sents()
paragraphsw = corpusw.paras()

wordse = corpuse.words()
sentencese = corpuse.sents()
paragraphse = corpuse.paras()

# Create nltk.Text objects from the words
textw = nltk.Text(wordsw)
texte = nltk.Text(wordse)
```

> * Define a function:
>     1. called `vocab_size(text)` that has a single parameter for the `text`, and which returns the vocabulary size of the text. Apply it to the rapper Wikipedia text;
>     2. `percent(word, text)` that calculates how often a given `word` occurs in a `text`, and expresses the result as a percentage. Apply it to the rapper Wikipedia Text to compute the percentage for *Dre*.

```python
def vocab_size(text):
    return len(set(word.lower() for word in text if word.isalpha()))

vocab_size(text)
```

```python
def percent(word, text):
    # Calculate the frequency distribution of the words in the text
    freq_dist = nltk.FreqDist(text)
    
    # Calculate the percentage of the word
    word_percent = (freq_dist[word] / len(text)) * 100
    
    return word_percent

percent("Dre", text)
```
