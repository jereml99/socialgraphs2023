> Use this method to retrieve a nice copy of all rappers' text. Save each rapper's text in a txt file and collect them all in a new folder!


```python
import requests
import os
import pandas as pd

df_rappers = pd.read_csv("../../files/Rappers.csv")
baseurl = "http://en.wikipedia.org/w/api.php?"

for rapper in df_rappers["WikipediaPageName"]:
    file_path = f"../data/clean/{rapper}.txt"
    
    # Check if the file already exists
    if os.path.exists(file_path):
        #print(f"File {rapper}.txt already exists. Skipping.")
        continue

    params = {
        "action": "query",
        "prop": "extracts",
        "exlimit": "1",
        "explaintext": "1",
        "format": "json",
        "titles": rapper
     }
    wikitext = requests.get(baseurl, params = params)
    wikijson = wikitext.json()
    num = list(wikijson['query']['pages'].keys())[0]
    text = wikijson['query']['pages'][num]['extract']
    try:
        with open(f"../data/clean/{rapper}.txt", "w", encoding="utf-8") as f:
            f.write(text)
    except Exception as e:
        print(f"Failed to save the page: {rapper}.")
```

> * What is a bigram? How does it relate to `collocations`. Explain in your own words.

*Answer*: A bigram is a sequence of two adjacent words in a text. Collocations are pairs or groups of words that appear together more frequently than would be expected by chance
>    1. Compute the collocations on the rapper Wikipedia Text.


```python
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

finder = BigramCollocationFinder.from_words(text)
finder.apply_freq_filter(3)  # Only keep bigrams that appear at least 3 times

# You can use different measures to rank the bigrams
bigram_measures = BigramAssocMeasures()
collocations = finder.nbest(bigram_measures.pmi, 10)  # Top 10 collocations

print(collocations)
```

> * Define a function:
>     1. called `vocab_size(text)` that has a single parameter for the `text`, and which returns the vocabulary size of the text. Apply it to the rapper Wikipedia text;


```python
def vocab_size(text):
    unique_tokens = set(text)
    return len(unique_tokens)

# Apply the function to the rapper Wikipedia text
vocabulary_size = vocab_size(text.tokens)
print(f"Vocabulary Size: {vocabulary_size}")
```

> 2. `percent(word, text)` that calculates how often a given `word` occurs in a `text`, and expresses the result as a percentage. Apply it to the rapper Wikipedia Text to compute the percentage for *Dre*.

```python
def percent(word, text):
    word_count = text.count(word)
    total_words = len(text)
    percentage = (word_count / total_words) * 100 if total_words > 0 else 0
    return percentage

# Apply the function to compute the percentage for 'Dre'
percentage_dre = percent("Dre", text.tokens)
print(f"Percentage of 'Dre': {percentage_dre}%")
```
